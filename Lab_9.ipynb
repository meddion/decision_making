```julia
using Plots, Random, LinearAlgebra
```

```julia
N = 25
points = rand(Float64, (N,2))

maxPoint, idx1 = findmax(points, dims=1); println("maxPoint: ", maxPoint)
minPoint, idx2 = findmin(points, dims=1); println("minPoint: ", minPoint)
dist = norm(maxPoint - minPoint); println("Distance between maxPoint & min Point: ", dist)

#display(idx1[1])
#display(len(points))
#points[setdiff(1:end, ), :]

scatter(points[:, 1], points[:, 2], markersize = 5)
scatter!(maxPoint[:, 1], maxPoint[:, 2], color="red")
scatter!(minPoint[:, 1], minPoint[:, 2], color="green")
```
