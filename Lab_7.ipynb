```julia
using Random, Distributions, Plots, DataFrames, CSV
N = 5

# Task1: Lambda = N
p = Poisson(N)
xs = 0:30
ys = pdf(p, xs)
CSV.write("poisson.csv", DataFrame(X = xs, Y = ys))
plot(xs, ys, line=3)
```

```julia
# Task 2: =N, =(N/2)^0.5
p = Normal(N, (N/2)^0.5)
s = 0.02
xs = -100:s:100
ys = pdf(p, xs)
CSV.write("normal.csv", DataFrame(X = xs, Y = ys))
plot(xs, ys, xlims=(-10, 20), line=3)
```

```julia
# Task 2: a=-N, b=N
p = Uniform(-N, N)
s = 0.02
xs = -2N:s:2N
ys = pdf(p, xs)
CSV.write("uniform.csv", DataFrame(X = xs, Y = ys))
plot(xs, ys, line=3)
```
